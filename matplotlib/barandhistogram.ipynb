What is the difference between a bar plot and a histogram in Matplotlib?

 bar plot:-A bar plot displays discrete data as bars

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

histogram

histogram:- a histogram displays continuous data as bars that represent the frequency of data points in a given range.

```python
import matplotlib.pyplot as plt

import numpy as np

data = np.random.normal(size=1000)

plt.hist(data, bins=30)

plt.show()
```
