plt.grid() function :- The plt.grid() function is used to add a grid to a plot. It takes an optional which parameter that specifies which gridlines to display (major or minor).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid() #if we remove this the plot will appear without grid

plt.show()
```
